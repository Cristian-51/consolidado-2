```python
# Funcion recursiva para dividir dos numeros
def dividir_recursivo(dividendo, divisor):
    if divisor == 0:
        return "Error: División por cero"
    if dividendo < divisor:
        return 0
    return 1 + dividir_recursivo(dividendo - divisor, divisor)

# Solicitar datos
dividendo = float(input("Ingresa el dividendo: "))
divisor = float(input("Ingresa el divisor: "))

resultado = dividir_recursivo(dividendo, divisor)
print(f"El resultado de la división es: {resultado}")
```
