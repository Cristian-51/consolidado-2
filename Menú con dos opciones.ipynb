```python
# Menú con dos opciones
# a. OPCION CONTAR CONSONANTES SIN VOCALES
# b. SALIR
def contar_consonantes_sin_vocales(palabra):
    consonantes = 'bcdfghjklmnpqrstvwxyz'
    palabra = palabra.lower()  # Convertir todo a minúsculas para comparación
    contador = 0

    for letra in palabra:
        if letra in consonantes:
            contador += 1

    return contador

def menu():
    while True:
        print("\nMenú de opciones:")
        print("a. Contar consonantes sin vocales")
        print("b. Salir")

        opcion = input("Elige una opción (a/b): ").lower()

        if opcion == 'a':
            palabra = input("Ingresa una palabra: ")
            print("Cantidad de consonantes sin vocales:", contar_consonantes_sin_vocales(palabra))
        elif opcion == 'b':
            print("Saliendo del programa...")
            break
        else:
            print("Opción no válida. Intenta de nuevo.")

menu()
```
